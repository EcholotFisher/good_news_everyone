```python
sp = spacy.load('en_core_web_sm')
doc = sp(text)
```

```python
newText =' '
for word in doc:
    if word.pos_ in ['ADJ', 'NOUN']:
        newText = ' '.join((newText, word.text.lower()))
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS).generate(newText)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
